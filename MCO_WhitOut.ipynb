## regresion lineal apartir del error cuadratico medio vectorial 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cargamos nuestro dataset
from sklearn.datasets import load_boston
```

```python
# vemos las caracteristicas del Dataset
boston = load_boston()
print(boston.DESCR)
```

### Implementamos formula (MCO): $\beta = (X^{T}X)^{-1}X^{T}Y$

```python
# creamos nuestras matrices queremos predecir el costo de la vivienda 
# apartir del numero de habitaciones
X = np.array(boston.data[:, 5])
Y = np.array(boston.target)

plt.scatter(X ,Y, alpha = 0.2)

# Añadimos columna de 1s para el termino independiente
X = np.array([np.ones(506) ,X]).T

#implementamos MCO (minimo cuadrado ordinario)
B = np.linalg.inv(X.T@X)@X.T@Y

#trazamos linea de regresion
plt.plot([4, 9], [B[0] + B[1] * 4, B[0] + B[1] * 9], c = "red", alpha = 0.8)

plt.show()
```
